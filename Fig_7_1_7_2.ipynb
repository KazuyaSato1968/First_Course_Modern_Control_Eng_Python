```python
import numpy as np
from control import matlab
from matplotlib import pyplot as plt
from scipy import arange 
```

```python
#システムパラメータを与える
A = np.array([[0.0, 1.0], [-6.0,-5.0]]) #行列A
b = np.array([[0.0], [1.0]]) #ベクトルb
c = np.array([[1.0, 0.0],[0.0,1.0]])#x_{1}とx_{2}をプロットするためにcを単位行列にする
d = np.array([[0.0],[0.0]])#Cに合わせてdを零ベクトルとする

sysP = matlab.ss(A, b, c, d) #システムの状態空間表現
```

```python
#システム行列Aの固有値を求める
eig, eig_vec = np.linalg.eig(A)
print("pole = ",eig)
```

```python
#時間変数の定義
t = np.arange(0, 10, 0.01) #0から10まで0.01刻み

#初期値を与える
x0 = np.array([[1.0],[0.0]]) #x(0) = [1,0]^{T}の場合
x00 = np.array([[0.0],[0.0]]) #x(0) = [0,0]^{T}の場合

#ステップ応答の計算
y1, t1 = matlab.step(sysP, t, x0) #x(0) = [1,0]^{T}の場合のステップ応答
y2, t2 = matlab.step(sysP, t, x00) #x(0) = [0,0]^{T}の場合のステップ応答
```

```python
# 図7.1のプロット
plt.plot(t1, y1[:,0], label = "x_1(t)") #x_{1}を抽出してプロット
plt.plot(t1, y1[:,1], label = "x_2(t)") #x_{2}を抽出してプロット
plt.xlim([0,10]) #横軸（時間軸）の範囲の指定
plt.ylim([-1.0, 1.0]) #縦軸の範囲の設定
plt.grid(color='gray') #罫線を灰色で表示
plt.xlabel("time   t[s]") #横軸のラベル表示
plt.ylabel("output") #縦軸のラベル表示
plt.legend() #凡例の表示
plt.show() #グラフの表示
```

```python
# 図7.2のプロット
plt.plot(t2, y2[:,0], label = "x_1(t)") #x_{1}を抽出してプロット
plt.plot(t2, y2[:,1], label = "x_2(t)") #x_{2}を抽出してプロット
plt.xlim([0,10]) #横軸（時間軸）の範囲の指定
plt.ylim([-1.0, 1.0]) #縦軸の範囲の設定
plt.grid(color='gray') #罫線を灰色で表示
plt.xlabel("time   t[s]") #横軸のラベル表示
plt.ylabel("output") #縦軸のラベル表示
plt.legend() #凡例の表示
plt.show() #グラフの表示
```
